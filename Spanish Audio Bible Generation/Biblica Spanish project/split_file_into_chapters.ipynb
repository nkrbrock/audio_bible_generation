```python
import os
import shutil
```

```python
# Specify the path to the input text file
input_file_path = "/content/Chinese_text.txt"

# Specify the path to the output directory
output_directory = "/content/Bible"
```

```python
# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Read the input text file
with open(input_file_path, "r") as file:
    for line in file:
        line = line.strip()
        if line:
            # Extract book, chapter, and text from each line
            book, chapter_text = line.split(" ", 1)
            chapter, text = chapter_text.split(":", 1)
            text = text[2:]

            # Format the file name and directory path
            file_name = f"{book}_{chapter}.txt"
            directory_path = os.path.join(output_directory, book)

            # Create the book directory if it doesn't exist
            os.makedirs(directory_path, exist_ok=True)

            # Write the chapter text to the file
            file_path = os.path.join(directory_path, file_name)
            with open(file_path, "a") as chapter_file:
                chapter_file.write(f"{text} ")
```
