```python
import os
from pydub import AudioSegment
```

```python
def split_audio(input_file, output_directory):
     # Create the output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            input_file = os.path.join(input_folder, filename)
            audio = AudioSegment.from_file(input_file)
            chunk_length = 50 * 1000

            chunk_count = len(audio) // chunk_length + 1

            for i in range(chunk_count):
                start_time = i * chunk_length
                end_time = (i + 1) * chunk_length

                chunk = audio[start_time:end_time]

                chunk_filename = f"{os.path.splitext(filename)[0]}_{i}.wav"
                output_path = os.path.join(output_directory, chunk_filename)

                chunk.export(output_path, format="wav")

                print(f"Chunk {i + 1} of {filename} saved to {output_path}")
```

```python
input_folder = "/content/audio"
output_directory = "/content/el_samples"
```

```python
split_audio(input_folder, output_directory)
```
