```python
import os
from pydub import AudioSegment

def split_audio_files(folder_path, output_folder):
    # Check if the output folder exists, create if necessary
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over audio files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.wav') or file_name.endswith('.mp3'):
            file_path = os.path.join(folder_path, file_name)
            
            # Load the audio file using PyDub
            audio = AudioSegment.from_file(file_path)

            time_length = 1 * 60 * 1000

            # Split the audio file into 1-minute chunks
            chunks = []
            for i in range(0, len(audio), time_length):
                chunk = audio[i:i + time_length]
                chunks.append(chunk)

            # Save each chunk as a separate file
            for idx, chunk in enumerate(chunks):
                name = file_name.split('.')
                output_file = f"{name[0]}_{idx+1}.wav"
                output_path = os.path.join(output_folder, output_file)
                chunk.export(output_path, format='wav')
                
                print(f"Saved chunk: {output_path}")

# Specify the folder containing the audio files
input_folder = '/content/audio_input'

# Specify the folder where the 1-minute chunks will be saved
output_folder = '/content/samples'

# Call the function to split the audio files
split_audio_files(input_folder, output_folder)
```
