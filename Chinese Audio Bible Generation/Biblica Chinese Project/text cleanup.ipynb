```python
import re
```

```python
def remove_markup(input_text):
    # Define the regular expression pattern to match the <char> tags and their content
    pattern = r'<char[^>]*>(.*?)<\/char>'
    # Use re.sub() to replace the matched pattern with the text within the <char> tags
    cleaned_text = re.sub(pattern, r'\1', input_text)
    return cleaned_text
```

```python
def condense_lines(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    condensed_lines = []
    current_verse = ''

    for line in lines:
        if re.match(r'^[A-Z]{3}\s\d+:\d+', line):
            if current_verse:
                condensed_lines.append(current_verse.strip())
            current_verse = line.strip()
        else:
            current_verse += line.strip()

    # Append the last verse to the condensed lines list
    if current_verse:
        condensed_lines.append(current_verse.strip())

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write('\n'.join(condensed_lines))
```

```python
with open('/content/2022 Chinese.txt', 'r') as file:
  lines = file.readlines()

lines
```

```python
new_text = []

for i in lines:
  new_text.append(remove_markup(i))
```

```python
with open('/content/Clean Chinese.txt', 'w') as file:
  for i in new_text:
    file.write(i)
```
