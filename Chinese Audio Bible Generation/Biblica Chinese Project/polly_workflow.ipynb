```python
for chapter in os.listdir(book_directory):
  file = os.path.join(book_directory, chapter)
  with open(file, 'r') as f:
    text = f.read()

  try:
    response = polly_client.synthesize_speech(VoiceId='Zhiyu',
                  OutputFormat='mp3',
                  Text = f'{text}',
                  Engine = 'neural')

    name = chapter[:-4]
    with open(f'/content/{book}/{name}.wav', 'wb') as file:
      file.write(response['AudioStream'].read())
  except:
    pass
```
